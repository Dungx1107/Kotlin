# companion object trong Kotlin là cách Kotlin tạo thành viên thuộc về class thay vì thuộc về object instance.

Hiểu đơn giản:

- companion object ≈ static trong Java

Nhưng Kotlin không có từ khóa static, nên dùng companion object.

```kotlin
class User {

    companion object name {
        var name1 :String = "Kotlin"
        fun hello() {
            println("kakka")
        }
    }

}
```

```kotlin
User.name.hello()
```
