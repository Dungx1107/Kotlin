```kotlin
class Quiz {
    companion object StudentProgress {
        val total: Int = 10
        var current: Int = 3
    }

    fun print() {
        println("${Quiz.current} of ${Quiz.total}")
    }
}
```

```kotlin
//Quiz.print() // sai vi ko phai la singleton // object
var kaka = Quiz()
kaka.print()
```

# 1. Thêm một thuộc tính tiện ích


Kotlin cho phép mở rộng class

-- Kotlin cho phép thêm function hoặc property vào class đã tồn tại mà không cần sửa class gốc.

- Đây gọi là Extension.

```kotlin
val Quiz.StudentProgress.progressText: String
    get() = "${Quiz.current} of ${Quiz.total}"
```

```kotlin
println(Quiz.progressText)
```

# 2. Thêm một hàm mở rộng


```kotlin
fun Quiz.StudentProgress.printProgressBar() {
    repeat(Quiz.current) {
        println("kaka")
    }
}
```

```kotlin
Quiz.StudentProgress.printProgressBar()
```

```kotlin
fun Quiz.hello(){
    println("hello")
}
```

```kotlin
//Quiz.hello() // sai vì Quiz.hello() yêu cầu extension của Companion object, nhưng bạn lại tạo extension của class Quiz.
// Quiz.hello() chỉ hoạt động nếu function thuộc companion object.

var aka = Quiz()
aka.hello()
```
