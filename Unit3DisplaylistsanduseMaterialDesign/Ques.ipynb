```kotlin
// Định nghĩa class Generic
class Question<T>(
    val questionText: String,
    val answer: T,
    val difficulty: String
) {
    fun printQuestion() {
        println("Câu hỏi: $questionText")
        println("Độ khó: $difficulty")
    }
}
```

```kotlin
fun main() {
    val q1 = Question<String>("Thủ đô Việt Nam là gì?", "Hà Nội", "Dễ")
    val q2 = Question<Boolean>("Kotlin có hay không?", true, "Vừa")
    val q3 = Question<Int>("1 + 1 = ?", 2, "Dễ")

    q1.printQuestion()
    q2.printQuestion()
    q3.printQuestion()
}
```

```kotlin
main()
```
