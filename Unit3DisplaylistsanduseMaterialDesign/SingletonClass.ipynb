Singleton trong Kotlin là một design pattern đảm bảo chỉ có duy nhất 1 instance của một class tồn tại trong toàn bộ ứng dụng và mọi nơi đều truy cập vào cùng instance đó.

Kotlin hỗ trợ Singleton rất trực tiếp bằng từ khóa object, nên bạn không cần viết code phức tạp như trong Java.

# Singleton là gì? (Ý tưởng cốt lõi)

Một class Singleton có 3 đặc điểm:

- Chỉ tạo được 1 object duy nhất

- Không cho phép tạo thêm object khác

- Có điểm truy cập toàn cục

Ví dụ thực tế:

Logger của app

Config của ứng dụng

Database manager

Cache manager

- Vì những thứ này chỉ cần 1 instance cho toàn hệ thống.

```kotlin
object DatabaseManger{
    fun connect(){
        println("connect to database")
    }
}
```

```kotlin
DatabaseManger.connect()
```
