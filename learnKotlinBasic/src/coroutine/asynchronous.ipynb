```kotlin
import kotlin.system.*
```

```kotlin
import kotlinx.coroutines.*

fun main1() {
    runBlocking {
        println("Weather forecast")
        launch {
            printForecast()
        }
        launch {
            printTemperature()
        }
    }
}


suspend fun printForecast() {
    delay(1000)
    println("Sunny")
}

suspend fun printTemperature() {
    delay(1000)
    println("30\u00b0C")
}
```

```kotlin
main1()
```

```kotlin
fun main2() {
    runBlocking {
        println("Weather forecast")
        printForecast()
        printTemperature()
    }
}
```

```kotlin
main2()
```

```kotlin
fun main3() {
    runBlocking {
// Giả sử printForecast() mất 3 giây để chạy
        launch {
            printForecast()  // Mất 3 giây
        }

// Dòng này sẽ chạy NGAY LẬP TỨC, không đợi 3 giây!
        println("Tôi chạy ngay sau launch, không cần đợi printForecast() xong")

        launch {
            printTemperature()  // Coroutine thứ hai
        }
// Code tiếp tục chạy mà không đợi printTemperature() xong
    }
}
```

```kotlin
main3()
```

```kotlin
fun main4() {
    val time = measureTimeMillis {  // do thoi gian chay
        runBlocking {
            println("Weather forecast")
            launch {
                printForecast()
            }
            launch {
                printTemperature()
            }
        }
    }
    println("Execution time: ${time / 1000.0} seconds")
}
```

```kotlin
main4()
```

## async() giống như launch() nhưng CÓ TRẢ VỀ KẾT QUẢ thông qua đối tượng Deferred<T>

```kotlin
fun main5() {
    runBlocking {
        println("Weather forecast")
        val forecast: Deferred<String> = async {
            getForecast()
        }
        val temperature: Deferred<String> = async {
            getTemperature()
        }
        println("${forecast.await()} ${temperature.await()}")
        println("Have a good day!")
    }
}

suspend fun getForecast(): String {
    delay(1000)
    return "Sunny"
}

suspend fun getTemperature(): String {
    delay(1000)
    return "30\u00b0C"
}
```

```kotlin
main5()
```
